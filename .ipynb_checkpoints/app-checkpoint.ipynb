```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from tensorflow import keras
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

#  1锔 Funkcja do ekstrakcji klatek co 2 sekundy
def extract_frames(video_path, interval=2):
    cap = cv2.VideoCapture(video_path)
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    frames = [] 
    
    frame_idx = 0
    while cap.isOpened():
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_idx)
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY))  # Konwersja na grayscale
        frame_idx += fps * interval  # Pobieranie co 2 sekundy
    
    cap.release()
    return frames

#  2锔 Obliczanie Optical Flow (przepywu optycznego)
def calculate_optical_flow(frames):
    flow_magnitudes = []
    
    for i in range(len(frames) - 1):
        flow = cv2.calcOpticalFlowFarneback(frames[i], frames[i + 1], None, 
                                            pyr_scale=0.5,  # Skala piramidy
                                            levels=3,       # Liczba poziom贸w piramidy
                                            winsize=15,     # Rozmiar okna
                                            iterations=3,   # Liczba iteracji
                                            poly_n=5,       # Rozmiar filtra wielomianowego
                                            poly_sigma=1.2, # Odchylenie standardowe filtra
                                            flags=0)
        magnitude, _ = cv2.cartToPolar(flow[..., 0], flow[..., 1])
        flow_magnitudes.append(np.mean(magnitude))  # rednia warto ruchu
    
    return np.array(flow_magnitudes)

#  3锔 Autoencoder do wykrywania anomalii w ruchu
def build_autoencoder(input_dim):
    input_layer = Input(shape=(input_dim,))
    encoded = Dense(16, activation='relu')(input_layer)
    encoded = Dense(8, activation='relu')(encoded)
    decoded = Dense(16, activation='relu')(encoded)
    output_layer = Dense(input_dim, activation='linear')(decoded)

    autoencoder = Model(input_layer, output_layer)
    autoencoder.compile(optimizer='adam', loss='mse')
    return autoencoder

#  4锔 Klasteryzacja wynik贸w K-Means
def cluster_scenes(motion_scores, n_clusters=2):
    data = motion_scores.reshape(-1, 1)
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(data)
    return labels  # 0 = nudne momenty, 1 = ciekawe momenty

#  5锔 Pena analiza wideo
def analyze_video(video_path):
    print(" Ekstrakcja klatek...")
    frames = extract_frames(video_path)

    print(" Obliczanie Optical Flow...")
    motion_scores = calculate_optical_flow(frames)

    print(" Trenowanie autoenkodera...")
    autoencoder = build_autoencoder(input_dim=1)
    autoencoder.fit(motion_scores.reshape(-1, 1), motion_scores.reshape(-1, 1), epochs=50, batch_size=8, verbose=0)

    print(" Wykrywanie anomalii...")
    reconstructed = autoencoder.predict(motion_scores.reshape(-1, 1))
    reconstruction_error = np.abs(motion_scores - reconstructed.flatten())

    print(" Klasteryzacja wynik贸w...")
    labels = cluster_scenes(reconstruction_error)

    # Tworzenie DataFrame dla Pandas
    df = pd.DataFrame({"czas (s)": np.arange(len(motion_scores)) * 2, "ruch": motion_scores, "anomalie": reconstruction_error, "etykieta": labels})
    
    # Wizualizacja wynik贸w
    plt.figure(figsize=(12, 6))
    plt.plot(df["czas (s)"], df["ruch"], label="Optical Flow", color="blue")
    plt.scatter(df["czas (s)"][df["etykieta"] == 1], df["ruch"][df["etykieta"] == 1], color="red", label="Ciekawy moment")
    plt.xlabel("Czas (s)")
    plt.ylabel("Ruch (rednia Optical Flow)")
    plt.legend()
    plt.title("Analiza ciekawych moment贸w w filmie")
    plt.show()

    return df

#  6锔 Uruchomienie analizy wideo
video_path = "C:\\Users\\pwdlp\\OneDrive\\Pulpit\\Vods\\kotSra.mp4"
df_results = analyze_video(video_path)
print(df_results)
```

```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np
import cv2
from deepface import DeepFace

class VideoClipEnv(gym.Env):
    def __init__(self, frames, video_path, image_shape):
        super(VideoClipEnv, self).__init__()

        # Przestrze akcji: 0 (pomi), 1 (zachowaj)
        self.action_space = spaces.Discrete(2)
        self.image_shape = image_shape
        self.observation_space = spaces.Box(low=0, high=255, shape=self.image_shape, dtype=np.uint8)
        self.frames = frames
        self.video_path = video_path

        # Stan pocztkowy
        self.current_step = 0
        self.selected_clips = []


    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        self.current_step = 0
        self.selected_clips = []

        observation = self._get_observation()
        info = {}  # Dodaj pusty sownik info
        return observation, info

    def step(self, action):
        """Wykonuje akcj i zwraca now obserwacj, nagrod oraz flag zakoczenia epizodu."""
        if self.current_step >= len(self.frames):
            return np.zeros(self.observation_space.shape), 0, True, False, {}  # Dodano truncated=False

        reward = self._calculate_reward(self.current_step) if action == 1 else 0

        if action == 1:
            self.selected_clips.append(self.current_step)

        self.current_step += 1
        done = self.current_step >= len(self.frames)
        
        print(f"(step) ->Step: {self.current_step}, Reward: {reward:.2f}, Selected Clips: {self.selected_clips}")
        print(f"(step) ->Emotions: {self._detect_emotion([self.frames[self.current_step]])}")

        return self._get_observation(), reward, done, False, {}  # Nowy format

    def _get_observation(self):
        if self.current_step >= len(self.frames):
            return np.zeros(self.observation_space.shape, dtype=np.float32)
    
        frame = self.frames[self.current_step]
        # Jeli chcesz zmienia rozmiar obrazu (np. na 64x64)
        # frame = cv2.resize(frame, (self.image_shape[1], self.image_shape[0]))
        frame = frame.astype(np.float32) / 255.0
    
        return frame

    def _calculate_reward(self, step):
        """Oblicza nagrod na podstawie emocji oraz ruchu w sekwencji klatek."""
    
        # Ustalona dugo analizowanego klipu (np. 30 klatek)
        clip_length = 30  
    
        # Pobierz indeksy klatek dla danego klipu (upewniamy si, 偶e nie wychodzimy poza zakres)
        start = max(0, step - clip_length // 2)
        end = min(len(self.frames), step + clip_length // 2)
    
        clip_frames = self.frames[start:end]  # Pobieramy fragment filmu
    
        # **1. Wykrywanie emocji**
        emotion_score = self._detect_emotion(clip_frames)
    
        # **2. Analiza dynamiki ruchu**
        motion_score = self._calculate_motion(clip_frames)
    
        print(f"(_calculate_reward) -> Motion score: {motion_score:.2f}, Emotion score: {emotion_score:.2f}")
    
        # **Nagroda jako kombinacja cech**
        # Mo偶emy czy emocje i ruch w jedn nagrod.
        # Przykad: Nagroda oparta na r贸wnych wagach emocji i ruchu, w zale偶noci od tego, kt贸re maj dla Ciebie wiksze znaczenie.
        reward = (0.5 * emotion_score) + (0.5 * motion_score)
    
        return reward
    
    

    def render(self, mode='human'):
        """Wizualizuje postp."""
        print(f"Step: {self.current_step}, Selected Clips: {self.selected_clips}")
        print(f"Emotions: {self._detect_emotion([self.frames[self.current_step]])}")

    def close(self):
        """Zamyka rodowisko."""
        pass
    
    #funkcje dodatkowe
    def _detect_emotion(self, clip_frames):
        """Analizuje emocje na twarzach w danym klipie wideo (konwersja grayscale -> RGB)."""
    
        emotion_scores = []
    
        for frame in clip_frames:
            try:
                # frame_rgb = cv2.cvtColor(frame, cv2.COLOR_GRAY2RGB)
    
                analysis = DeepFace.analyze(frame, actions=['emotion'], enforce_detection=False)
    
                dominant_emotion = analysis[0]['dominant_emotion']  # np. 'happy', 'surprise', 'angry'
                
                #  Przyznawanie punkt贸w w zale偶noci od wykrytej emocji
                emotion_intensity = {
                    'happy': 1.0,
                    'surprise': 0.9,
                    'angry': 0.7,
                    'sad': 0.4,
                    'neutral': 0.2
                }
                
                emotion_scores.append(emotion_intensity.get(dominant_emotion, 0))
            except:
                emotion_scores.append(0)  # Brak twarzy na klatce
        
        return np.mean(emotion_scores) if emotion_scores else 0
    
    def _calculate_motion(self, clip_frames):
        """Oblicza poziom ruchu na podstawie zmian midzy klatkami."""
    
        motion_scores = []
    
        for i in range(1, len(clip_frames)):
            frame1 = clip_frames[i - 1]
            frame2 = clip_frames[i]
    
            # Optical Flow - analiza r贸偶nic ruchu midzy klatkami
            flow = cv2.calcOpticalFlowFarneback(frame1, frame2, None, 0.5, 3, 15, 3, 5, 1.2, 0)
    
            # Mierzymy redni prdko ruchu na ekranie
            motion_intensity = np.linalg.norm(flow)
            motion_scores.append(motion_intensity)
    
        return np.mean(motion_scores) if motion_scores else 0
    
    def _generate_final_video(self, output_path="C:\\Users\\pwdlp\\OneDrive\\Pulpit\\Vodsselected_clips.mp4"):
        """Tworzy finalny film na podstawie wybranych klatek."""
        
        if not self.selected_clips:
            print("锔 Brak wybranych klatek! Nie mo偶na wygenerowa filmu.")
            return
    
        # Otwieramy oryginalne wideo
        cap = cv2.VideoCapture(self.video_path)
        fps = int(cap.get(cv2.CAP_PROP_FPS))
    
        height, width = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)), int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))
    
        # Dodajemy wybrane klatki do nowego filmu
        for frame_idx in self.selected_clips:
            cap.set(cv2.CAP_PROP_POS_FRAMES, frame_idx)
            ret, frame = cap.read()
            if ret:
                out.write(frame)
    
        cap.release()
        out.release()
        print(f" Film zapisano jako: {output_path}")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from stable_baselines3 import A2C
import cv2
from gymnasium.wrappers import RecordEpisodeStatistics

#wychodzi na to ze na razie git (jednak nie byo git)
#trzeba rozkminic jak jest z tym fps czy dobrze te klatki w sensie czasu

def extract_frames(video_path, interval=2):
    cap = cv2.VideoCapture(video_path)
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    
    # Pobranie wymiar贸w pierwszej klatki
    ret, first_frame = cap.read()
    if not ret:
        cap.release()
        raise ValueError("Nie mo偶na odczyta pierwszej klatki wideo!")

    height, width, channels = first_frame.shape  # Pobranie wymiar贸w obrazu
    
    # Konwersja pierwszej klatki na grayscale i dodanie trzeciego wymiaru
    first_frame_gray = cv2.cvtColor(first_frame, cv2.COLOR_BGR2GRAY)
    frames = [np.expand_dims(first_frame_gray, axis=-1)]  # dodajemy kana, by mie (wysoko, szeroko, 1)

    frame_idx = fps * interval  # Pierwsza klatka ju偶 dodana
    while cap.isOpened():
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_idx)
        ret, frame = cap.read()
        if not ret:
            break
        # Konwersja do grayscale i dodanie kanau
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        frames.append(np.expand_dims(gray_frame, axis=-1))  # dodajemy kana, by mie (wysoko, szeroko, 1)
        frame_idx += fps * interval
    
    cap.release()
    return frames, (height, width, 1)


# Przykadowy model obrazu (CNN) DO WYJEBANIAAAAAA chyba ze wyszpcimy model do ekstrakcji cech z obrazu
def build_image_model(input_shape):
    inputs = layers.Input(shape=input_shape)
    x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(inputs) 
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Flatten()(x)
    x = layers.Dense(64, activation='relu')(x)
    return models.Model(inputs, x, name="image_model")


# Budowanie modeli
video_path = "C:\\Users\\pwdlp\\OneDrive\\Pulpit\\Vods\\kotSra.mp4"

frames, image_shape = extract_frames(video_path)
# frames_resized = [cv2.resize(frame, (64, 64)) for frame in frames]
# image_model = build_image_model(image_shape)

env = VideoClipEnv(frames, video_path, image_shape)
env = RecordEpisodeStatistics(env)

model = A2C('MlpPolicy', env, verbose=1, n_steps=4, ent_coef=0.01, learning_rate=0.0003, gamma=0.99)



model.learn(total_timesteps=100)

# Testowanie modelu
obs, info = env.reset()
for _ in range(len(frames)):
    print(f"Before step: Step {_}")
    action, _states = model.predict(obs, deterministic=True)
    obs, reward, done, _, info = env.step(action)
    print(f"After step: Step {_}, Reward: {reward}, Done: {done}")
    env.render()
    if done:
        break

env.close()
```
